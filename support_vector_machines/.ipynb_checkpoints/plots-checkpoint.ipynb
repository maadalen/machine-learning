```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_1 = [3, 3, 6, 6]
Y_1 = [1, -1, 1, -1]
X_2 = [1, 0, 0, -1]
Y_2 = [0, 1, -1, 0]

fig, ax = plt.subplots()

ax.scatter(X_1, Y_1, color = 'r', label = "Positive Labels")
ax.scatter(X_2, Y_2, color = 'b', label = "Negative Labels")
fig.suptitle('Plot of Training Data and Labels', fontsize=12, fontweight='bold')
ax.axvline(x= 2, color = 'k', label = "Maximal Margin Hyperplane")

legend = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

frame = legend.get_frame()
frame.set_facecolor('0.90')

for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)  # the legend line width


plt.savefig('Scatter_Plot.pdf', format="pdf")
plt.show()
```

```python
X_1 = [-1, -1]
Y_1 = [0, 1]
X_2 = [2,3,4]
Y_2 = [0, 0, 0]

fig, ax = plt.subplots()

ax.scatter(X_1, Y_1, color = 'r', label = "Positive")
ax.scatter(X_2, Y_2, color = 'b', label = "Negative")
fig.suptitle('Plot of Training Data and Labels', fontsize=12, fontweight='bold')
ax.axvline(x= 0.5, color = 'k', label = "Max Hyperplane")

legend = ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

frame = legend.get_frame()
frame.set_facecolor('0.90')

for label in legend.get_texts():
    label.set_fontsize('large')

for label in legend.get_lines():
    label.set_linewidth(1.5)  # the legend line width

plt.savefig('Scatter_Plot.pdf', format="pdf")
plt.show()
```
